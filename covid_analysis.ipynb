```python
dataset1= pd.read_csv("covid.csv")
dataset2= pd.read_csv("covid_grouped.csv")
```

```python
dataset1['ActiveCases'] = (
    dataset1['TotalCases']
    - dataset1['TotalDeaths']
    - dataset1['TotalRecovered']
)
```

```python
dataset2['Date'] = pd.to_datetime(dataset2['Date'])
```

```python
px.bar(dataset1.head(15), x = 'Country/Region',
       y = 'TotalCases',color = 'TotalCases',
       height = 500,hover_data = ['Country/Region', 'Continent'])
```

```python
top15 = dataset1.sort_values('TotalCases', ascending=False).head(15)


fig = px.bar(top15,
             x='Country/Region',
             y='TotalCases',
             color='TotalCases',
             hover_data=['Country/Region', 'Continent'],
             height=500,
             title='Top 15 Countries by Total COVID Cases')

fig.show()

```

```python
px.bar(dataset1.head(15),
       x='TotalTests',
       y='Continent',
       color='TotalTests',
       orientation='h',
       height=500,
       hover_data=['Country/Region', 'Continent'])
```

```python


fig = px.scatter(dataset1,
                 x='Continent',
                 y='TotalCases',
                 size='TotalCases',
                 color='Continent',
                 hover_data=['Country/Region', 'Continent'],
                 height=600,
                 title='Continent-wise Total COVID Cases')

fig.show()
```

```python
px.bar(dataset2, x="Date", y="Confirmed", color="Confirmed",
       hover_data=["Confirmed", "Date", "Country/Region"], height=400)
```

```python
px.scatter(dataset2, x="Confirmed", y="Deaths", height=400)
```

```python
px.choropleth(dataset2,
              locations="iso_alpha",
              color="Confirmed",
              hover_name="Country/Region",
              color_continuous_scale="Blues",
              animation_frame="Date")
```
